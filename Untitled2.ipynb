```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# --- Original Data ---
data = {
    'age': [25, 30, None, 20, None, 28],
    'salary': [50000, None, 60000, 52000, 50000, None]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)

# --- Mean Imputation ---
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("\nImputed Data:")
print(df_imputed)

# --- Plot Before vs After Imputation ---
plt.figure(figsize=(12,5))

# After Imputation
plt.subplot(1, 2, 2)
plt.title("After Mean Imputation")
plt.scatter(df_imputed.index, df_imputed["age"], label='Age', color='blue')
plt.scatter(df_imputed.index, df_imputed["salary"], label='Salary', color='red')
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
